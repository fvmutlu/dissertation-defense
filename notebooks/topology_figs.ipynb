```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
dim = 4
xx, yy = (dim,dim)
x = np.linspace(0,1,xx)
y = np.linspace(0,1,yy)
xv, yv = np.meshgrid(x,y)
pos = {}
node_labels = {}
for i in range(dim):
    for j in range(dim):
        pos[(i,j)] = [xv[i][j],yv[i][j]]
        node_labels[(i,j)] = f"{i*dim+j}"
```

```python
G = nx.grid_2d_graph(dim,dim)
fig, ax = plt.subplots(figsize=(3,3))
nx.draw(
    G, pos, ax=ax, edge_color='black', width=2, linewidths=2,
    node_size=500, node_color='red', alpha=1, font_size=14
)
plt.savefig("../images/grid.svg",bbox_inches='tight',pad_inches=0)
```
